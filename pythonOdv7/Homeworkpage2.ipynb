```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('BostonHousing.csv')
```

```python
print(df.head())
```

```python
# GRAFK 1 - YA VE EV MEDYAN DEER
plt.figure(figsize=(10, 6))
plt.scatter(df['age'], df['medv'], color='blue', alpha=0.7)
plt.xlabel("age (Ev Ya)", fontsize=12)
plt.ylabel("medv (Ev medyan deeri)", fontsize=12)
plt.title('Ev ya ve ev medyan deeri arasndaki iliki')
plt.grid(True)
plt.show()
```

```python
# GRAFK 2 - ODA SAYISI VE EV MEDYAN DEER
plt.figure(figsize=(10, 6))
plt.scatter(df['rm'], df['medv'], color='green', alpha=0.7)
plt.xlabel('rm (Oda Says)', fontsize=12)
plt.ylabel('medv (Ev Medyan Deeri)', fontsize=12)
plt.title('Oda Says ve Ev Medyan Deeri Arasndaki liki')
plt.grid(True)
plt.show()
```
