```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('BostonHousing.csv')
```

```python
print(df.head())
```

```python
# GRAFİK 1 - YAŞ VE EV MEDYAN DEĞERİ
plt.figure(figsize=(10, 6))
plt.scatter(df['age'], df['medv'], color='blue', alpha=0.7)
plt.xlabel("age (Ev Yaşı)", fontsize=12)
plt.ylabel("medv (Ev medyan değeri)", fontsize=12)
plt.title('Ev yaşı ve ev medyan değeri arasındaki ilişki')
plt.grid(True)
plt.show()
```

```python
# GRAFİK 2 - ODA SAYISI VE EV MEDYAN DEĞERİ
plt.figure(figsize=(10, 6))
plt.scatter(df['rm'], df['medv'], color='green', alpha=0.7)
plt.xlabel('rm (Oda Sayısı)', fontsize=12)
plt.ylabel('medv (Ev Medyan Değeri)', fontsize=12)
plt.title('Oda Sayısı ve Ev Medyan Değeri Arasındaki İlişki')
plt.grid(True)
plt.show()
```
